This means the algorithm takes small steps in the direction of the negative gradient to minimize the function. The process continues until one of two stopping conditions is met: either the change between consecutive steps becomes smaller than

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.cos(2 * np.pi * x) + x**2

def f_prime(x):
    return -2 * np.pi * np.sin(2 * np.pi * x) + 2 * x
alpha = 0.05
max_iters = 60
tol = 1e-6
x = 1.5
path = [x]
for _ in range(max_iters):
    grad = f_prime(x)
    new_x = x - alpha * grad
    path.append(new_x)
    if abs(new_x - x) < tol:
        break
    x = new_x
xs = np.linspace(-2, 2, 400)
ys = f(xs)
plt.figure(figsize=(8, 5))
plt.plot(xs, ys, label="f(x) = cos(2œÄx) + x¬≤")
plt.plot(path, [f(p) for p in path], marker='o', linestyle='-', label="Gradient‚Äëdescent path")
plt.title("Gradient¬†Descent on f(x) = cos(2œÄx) + x¬≤")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
